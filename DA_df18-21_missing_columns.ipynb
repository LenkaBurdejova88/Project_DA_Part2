# Find missing columns in dataframes

```python
import pandas as pd

df18 = pd.read_csv("df18_final_headers.csv")

df19 = pd.read_csv("df19_final_headers.csv")

df20 = pd.read_csv("df20_final_headers.csv")

df21 = pd.read_csv("df21_final_headers.csv")


# DataFrames to list
dfs = [df18, df19, df20, df21]
df_names = ["df18", "df19", "df20", "df21"]

all_columns = set().union(*[set(df.columns) for df in dfs])

# for each Dataframe we find which columns are missing
for df, name in zip(dfs, df_names):
    missing = all_columns - set(df.columns)
    if missing:
        print(f"Missing columns {name}:")
        for col in missing:
            print(f" - {col}")
    else:
        print(f"Missing columns {name}: None")
```
