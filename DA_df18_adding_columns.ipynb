Insert missing columns

* list of missing columns from df18-21_missing_columns.ipynb

```python
missing_columns_text="""
Machine_Learning_Frameworks_Huggingface
Data_Science_Courses_Cloud_Certification_Programs
Notebook_Products_Aws_Notebook
Data_Science_Courses_Linkedin_Learning
Notebook_Products_Zeppelin
Notebook_Products_Amazon_Sagemaker_Studio
Notebook_Products_Zepl_Notebooks
Programming_Languages_Swift
Notebook_Products_Ibm_Watson_Studio
Notebook_Products_Amazon_Emr_Notebooks
Machine_Learning_Frameworks_Tidymodels
Notebook_Products_Code_Ocean
Machine_Learning_Frameworks_Pytorch_Lightning
Notebook_Products_Deepnote_Notebooks
Notebook_Products_Google_Cloud_Datalab_Notebooks
Machine_Learning_Frameworks_Jax
Notebook_Products_Observable_Notebooks
Notebook_Products_Databricks_Collaborative_Notebooks"""

missing_columns = missing_columns_text.strip().splitlines()

print(missing_columns)

len(missing_columns)
```

```python
#fill bullets with NaN

for column in missing_columns:
    df18[column] = np.nan

df18
```

```python
# check all columns
df18.to_csv("df18_col_complete.csv", index=False)
```

Rearrangement of column order

```python
first_columns = ["Year", "Age", "Country", "Education", "Gender",
                 "Current_Role","Current_Yearly_Compensation", "Time_Writing_Code",
                  "Time_Using_Machine_Learning_Methods",
                  "Primary_Tool_Used_For_Data_Analysis"]

# all columns in df
all_columns = df18.columns.tolist()

# columns which aren't in first_columns
residue = sorted([col for col in all_columns if col not in first_columns])

# new order
new_order = first_columns + residue

# print(new_order)

#rearrangement of columns in df
df18 = df18[new_order]

#save as csv file
df18.to_csv("df18_all_col.csv", index=False)
```
